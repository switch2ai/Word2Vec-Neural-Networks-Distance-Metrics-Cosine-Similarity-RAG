##Similarity Metric :

```python
1) Manhatten distence : |X2 - X1 | + | Y2 - Y1 |
    >> Less the distance more the similarity

2) Euclidian distance : sqrt( (X2 - X1)^2 + (Y2 - Y1)^2 )
    >> Less the distance more the similarity

3) Dot Product :|a|||b|cosθ
    >> Larger the value more the similarity
    >> Less the value (near to 0) less the similarity

4) Cosine Similarity : cosθ
    >> Larger the value (closer to 1) more the similarity
    >> Less the value (near to 0) less the similarity

## Why cosine similarity and not sine or tan?
      -- Aligned with human intuation
      -- range is fixed, easy to interpret
```
